# Data: Argelaguet et al.

In this notebook, we combine the count table with the meta data and filter for embryonic day 5,5.

```python
import scanpy as sc
import anndata
```

```python
meta_argelaguet = sc.read_text('../data/rnaseq_mouse_e55/Argelaguet_meta.txt', dtype="str", delimiter="\t")
adata_argelaguet = sc.read("../data/rnaseq_mouse_e55/Argelaguet_counts.h5ad")
print(meta_argelaguet)
print(adata_argelaguet)
```

```python
adata_argelaguet.to_df()
```

### Add metadata to obs

```python
for var_name in meta_argelaguet.var_names:
    adata_argelaguet.obs[var_name] = meta_argelaguet.to_df()[var_name]
```

```python
adata_argelaguet.obs[:]
```

### Filter for embryonic day 5.5

```python
adata_argelaguet_E55 = adata_argelaguet[adata_argelaguet.obs["stage"] == "E5.5"]
```

```python
adata_argelaguet_E55
```

Unfortunatly all celltypes at stage E5.5 are NAs!

```python
adata_argelaguet_E55.obs["celltype"].unique()
```

### Write the filtered file

```python
adata_argelaguet_E55.write_h5ad("../data/rnaseq_mouse_e55/Argelaguet_E55.h5ad")
```
